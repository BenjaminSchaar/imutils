```python
# install the MicroscopeDataReader through imutils
# https://github.com/Zimmer-lab/imutils
# read the Docs of imutils!
from imutils import MicroscopeDataWriter
inspect(MicroscopeDataWriter)
"""
Args:
    dataset_path (Path or str): Path to the file or directory writing the data
     (str): Name of the dataset (folder name)
    add_date_time (bool, optional): Add date and time to the dataset name. Defaults to True.
    summary_metadata (dict, optional): Summary metadata for the dataset. Defaults to None.
    verbose (int, optional): Verbosity level of logger messages. Defaults to 1, all usual logging messages. 0 for no logging messages. 2 for extended debugging messages.
    **kwargs: Additional arguments for the NDTiffDataset class
"""
```

```python
# Writing Data with MicroscopeDataWriter
import numpy as np # to generate random data
import time

summary_metadata = {'meta_key_1': 123, 'meta_key_2': 'something else'} # dictionary, make this whatever you want
# the metadata is also written as json, thefore only use (str, int, float, bool, None) types.

# open dataset without metadata
dataset = MicroscopeDataWriter(dataset_path="P:\TestData", dataset_name="FirstNdtiffTest", summary_metadata=summary_metadata)

image_height = 256
image_width = 256

# write 10 images
for t in range(10):
    # generate random image
    pixels = np.random.randint(0, 2 ** 16, (image_height, image_width), dtype=np.uint16)
    time.sleep(0.1) # to see that the time metadata works
    image_metadata = {'name_a': t, 'name_v': 'something'} # whatever you want
    # the metadata is also written as json, thefore only use (str, int, float, bool, None) types.

    # put image into the dataset (gets written asynchronous)
    dataset.put_image(image=pixels, time=t, image_metadata=image_metadata)
```

```python
inspect(dataset, methods=True)
```

```python
inspect(dataset.get_image_coordinates_list())
inspect(dataset.get_index_keys())
inspect(dataset.get_summary_metadata())
inspect(dataset.read_image(time=0))
inspect(dataset.read_image_metadata(time=0))
```
