```python
import cv2
import matplotlib.pyplot as plt
import os
import glob
import tifffile as tiff
```

```python
#Saves the first frame of the mp4 files in a folder

path='/groups/zimmer/Ulises/Conditioning/Chemotaxis/201907_and_201908/together/mp4/'

output_path='/groups/zimmer/Ulises/Conditioning/Chemotaxis/201907_and_201908/together/first_frame/'

for directory, subdirectories, files in os.walk(path):
    print('There are ', len(files), ' files:')
    for file in files:
        print(file)
        filepath=os.path.join(directory,file)
        print(filepath)
        cap = cv2.VideoCapture(filepath)
        ret, frame = cap.read()
        filename=os.path.basename(file).split('.')[0]
        save_path=os.path.join(output_path,filename+'.tiff')
        print(save_path)
        tiff.imwrite(save_path,frame)
```
